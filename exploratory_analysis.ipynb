**1. Importei as bibliotecas necessárias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**2. Importei os dados**

```python
data = pd.read_csv('data_clean.csv')
data_numeric = pd.read_csv('data_num.csv')
data_numeric.head(10)
```

**3. Verifiquei a distribuição dos dados de cadacoluna utilizando histogramas**

```python
fig, ax = plt.subplots(4, 2, figsize=(15, 15))
sns.set_style('whitegrid')
for variable, subplot in zip(data_numeric.columns, ax.flatten()):
    sns.distplot(data_numeric[variable], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
```

**4. Verifiquei por relações entre as variáveis utilizando heatmap**

```python
corr = data_numeric.corr()
fig, ax = plt.subplots(figsize=(25, 25), dpi=100)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
```

**5. Verifiquei por relações entre as variáveis utilizando pairplot**

```python
sns.pairplot(data_numeric)
```
